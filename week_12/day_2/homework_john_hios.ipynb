**Q3.** Select the variables `Beverage_category`, `Beverage`, `Beverage prep` and `Calories` from the `drinks_content` data frame, and assign the selected columns to a new data frame called `drinks`. Check if there are any missing  values anywhere in the data, and drop any rows containing them if there are.

```python
# we use .copy() at the end so we can disentangle drinks from drinks_content

drinks = drinks_content.loc[:,["Beverage_category", "Beverage", "Beverage_prep", "Calories"]].copy()
```

**Q4.** In your `drinks` dataframe create a new variable (column) called `calorie_diff`, which stores the difference between 135 calories (135 calories = 10 cubes of sugar!) and the calories in each drink. (Hint: you’ll want to subtract the calories from 135 to see which drink have more or less than 10 cups of sugar).

```python
drinks["calorie_diff"] = 135 - drinks["Calories"]

drinks.head()
```

**Q5.** Summarise the mean number of calories in each beverage category.

```python
drinks.groupby("Beverage_category")["Calories"].mean().reset_index(name = "mean_calories")
```

**Q6.** **Harder** - Which `Beverage_prep` type contains more than the average calories of all drinks?
(Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
# calculate the average calories

avg_calories = drinks["Calories"].mean()

avg_calories
```

```python
# get the grouped Beverage_prep data

beverage_prep_df = drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_calories").copy()
```

```python
# find which prep types have above average calories

beverage_prep_df.loc[beverage_prep_df["mean_calories"] > avg_calories]
```

**Q7.** **Harder**  - Which is the best type of **coffee drink** to get if you’re worried about consuming too many calories?  (Hint: you can use pd.unique() on a column to get all unique values/levels of that column! Also remember that you can use isin() to subset with a list)

```python
# can start by finding the drink with the lowest number of calories

drinks_min_calories = drinks.groupby(["Beverage_category", "Beverage", "Beverage_prep"])["Calories"].min().reset_index(name = "min_calories").copy()

drinks_min_calories.sort_values("min_calories", ascending = True).head(15)
```

```python
# get unique drinks categories
pd.unique(drinks_min_calories["Beverage_category"])
```

```python
# filter out the Tazo tea drinks to get the lowest calorie coffee drink.

coffee_drinks = drinks.loc[drinks["Beverage_category"].isin(['Coffee', 
                                                             'Classic Espresso Drinks', 
                                                             'Signature Espresso Drinks',
                                                             'Frappuccino® Blended Coffee', 
                                                             'Frappuccino® Light Blended Coffee'])].copy()
coffee_drinks
```

```python
coffee_drinks.sort_values("Calories").head(1)
```

**Q8.** Filter the dataset drinks so we only take "Classic Espresso Drinks", and save this in a new data frame called `espresso_drinks`.

```python
espresso_drinks = drinks.loc[drinks["Beverage_category"] == "Classic Espresso Drinks"].copy()

espresso_drinks.head()
```

**Q9.** Group your `espresso_drinks` data frame by the type of `Beverage prep`, and then find out the mean calories for each drink group.

```python
espresso_drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_calories")
```

**Q10.** Get the same grouped mean `Calories` values as above, but this time sorted in descending order. (Hint: so you're able to sort the resulting groupby by the 'Calories' column you will want to ensure you get back a dataframe and not a series)

```python
espresso_drinks.groupby("Beverage_prep")["Calories"].mean().sort_values(ascending = False).reset_index(name = "mean_calories")
```
